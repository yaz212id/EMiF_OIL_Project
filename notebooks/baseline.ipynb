```python
import pandas as pd
import numpy as np
import statsmodels.api as sm

df = pd.read_csv("../data/interim/daily_clean.csv")
df['Date'] = pd.to_datetime(df['Date'])


# S&P500 returns
df['r_oil'] = np.log(df['Brent futures'] / df['Brent futures'].shift(1))
df['r_sp500'] = np.log(df['S&P500'] / df['S&P500'].shift(1))

df['r_oil_pos'] = df['r_oil'].clip(lower=0)

df['r_sp500_lag'] = df['r_sp500'].shift(1)


df = df.dropna()


X_ret = df[['r_oil_pos', 'r_sp500_lag']]
X_ret = sm.add_constant(X_ret)

y_ret = df['r_sp500']

model_ret = sm.OLS(y_ret, X_ret).fit()
print(model_ret.summary())
```

```python
# S&P500 volatility
df['vol_sp500'] = df['r_sp500']**2
df['vol_oil'] = df['r_oil']**2


df['vol_sp500_lag'] = df['vol_sp500'].shift(1)
df['vol_oil_lag'] = df['vol_oil'].shift(1)

df_vol = df.dropna()



X_vol = df_vol[['vol_oil', 'vol_oil_lag', 'vol_sp500_lag']]
X_vol = sm.add_constant(X_vol)

y_vol = df_vol['vol_sp500']

model_vol = sm.OLS(y_vol, X_vol).fit()
print(model_vol.summary())
```

```python
# Gold returns
df['r_gold'] = np.log(df['Gold'] / df['Gold'].shift(1))

df['r_gold_lag'] = df['r_gold'].shift(1)


df = df.dropna()


X_ret = df[['r_oil_pos', 'r_gold_lag']]
X_ret = sm.add_constant(X_ret)

y_ret = df['r_gold']

model_ret = sm.OLS(y_ret, X_ret).fit()
print(model_ret.summary())
```

```python
# Gold volatility
df['vol_gold'] = df['r_gold']**2


df['vol_gold_lag'] = df['vol_gold'].shift(1)

df_vol = df.dropna()


X_vol = df_vol[['vol_oil', 'vol_oil_lag', 'vol_gold_lag']]
X_vol = sm.add_constant(X_vol)

y_vol = df_vol['vol_gold']

model_vol = sm.OLS(y_vol, X_vol).fit()
print(model_vol.summary())
```
